# Pandas Data Visualization
- Only learn this lecture once you read matplotlib and seaborn

```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# read the first data frame csv and set the index to be the first column
df1 = pd.read_csv("df_files/df1.csv", index_col=0)

df1.head()
```

```python
# read the second data frame csv
df2 = pd.read_csv("df_files/df2.csv")

df2.head()
```

### Methods of Plotting using Pandas

```python
"""
create a histogram for the A column of df1
Note: histograms a type of graph that show the frequency,
which is the number of times a value occurred
"""
df1["A"].hist()
```

```python
import seaborn as sns

"""
create a histogram with 25 bins for the A column of
df1 with seaborn overriding the styles
"""
df1["A"].hist(bins=25)
```

```python
# another way of creating a histogram, by plotting
df1["A"].plot(kind="hist", bins=25)
```

```python
"""
and the last and best way creating a histogram
- ec="black" creates black outlines on each x-axis
"""
df1["A"].plot.hist(ec="black")
```

### Different Types of Pandas Plots

```python
# shows a plot of area of the values, like a line plot but filled colors
df2.plot.area()
```

```python
# a bar plot with values stacked onto each other instead of separated
df2.plot.bar(stacked=True)
```

```python
# plot the values of the B column in respect to the index (date)
df1.plot.line(use_index=True, y="B", figsize=(12, 3))
```

```python
"""
plot the values of B in respect to A as a scatter plot
- c="C" means to also plot C but as a different color
"""
df1.plot.scatter(x="A", y="B", c="C")
```

```python
"""
plot the values of B in respect to A as a scatter plot
- s=df1["C"] * 100 means to also plot C but as a size * 100
"""
df1.plot.scatter(x="A", y="B", s=df1["C"] * 100)
```

```python
# plot a box plot for each column's values
df2.plot.box()
```

```python
# create a DataFrame of random values of 1000 rows and 2 columns
rand_df = pd.DataFrame(np.random.randn(1000, 2), columns=["a", "b"])

rand_df.head()
```

```python
# create a hexbin plot of column b in respect to column a
rand_df.plot.hexbin(x="a", y="b", gridsize=25)
```

```python
"""
plot the kernel density estimation (KDE) of all columns
- can also use df2.plot.density()
"""
df2.plot.kde()
```
