# Seaborn Style and Color

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read the tips sample data set
tips = sns.load_dataset("tips")

tips.head()
```

```python
# set the background style of the figure
sns.set_style("whitegrid")

# set the aspect ratio of the figure to 12x3
plt.figure(figsize=(12, 3))

"""
set the figure context as a poster with font scaling of 0.75

Check out the other figure contexts here:
- https://seaborn.pydata.org/generated/seaborn.set_context.html
"""
sns.set_context("poster", font_scale=0.75)

# count plot the sex
sns.countplot(x=("sex"), data=tips)

# (must go after plot) remove left, right, top, and bottom spines of figure
sns.despine(left=True, right=True, top=True, bottom=True)
```

```python
"""
plot the tip by total_bill with hue as sex with a palette of seismic

Check out all the other color maps here:
- https://matplotlib.org/tutorials/colors/colormaps.html
"""
sns.lmplot(x=("total_bill"), y=("tip"),
           data=tips, hue=("sex"), palette=("seismic"))
```
