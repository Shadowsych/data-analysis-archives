# Seaborn Distribution Plots
Show a uni-variant (one variable) set of observations as a plot.
- Uses a kernel density estimation (KDE) line
- Shows a distribution of bars for each data

```python
# import the seaborn module
import seaborn as sns
```

```python
# visualize matplotlib inside the notebook
%matplotlib inline
```

```python
# sample pandas DataFrame from seaborn for restraunt tips
tips = sns.load_dataset("tips")

# read a few rows of the data set
tips.head()
```

```python
"""
show a distribution plot (like a histogram) for the total bill
- for this graph, do not show a kde line (kde=False)
- bins refers to the number of distributions to show, increasing bins
would mean that you would sooner or later plot every single item
"""
sns.distplot(tips["total_bill"], kde=False, bins=10)
```

### JointPlot
Matches two distribution plots for an x, y

```python
"""
create a jointplot (match up two distribution plots) as an x, y
- create a regression-kind of distribution instead of the default scatter
Note: there are many "kinds", but they all essentially show the same data

If you look at the graph below, the vertical distribution plot is for
the total_bill and the horizontal distribution plot is for the tip.
Together, they create a regression plot.

See more about jointplot here:
- https://seaborn.pydata.org/generated/seaborn.jointplot.html
"""
joint = sns.jointplot(x=("total_bill"), y=("tip"), data=tips, kind="reg")
```

### PairPlot
Create jointplots for all combinations of numerical columns in a DataFrame

```python
"""
create a pair plot for each numerical column combination in tips

- hue allows us to separate the datas with a legend and colors,
so for this data set we separate them by sex
- palette allows us to change the color of the plots
"""
sns.pairplot(tips, hue=("sex"), palette="coolwarm")
```

### RugPlot
Draws a dash mark for every point on a uni-variant (one variable) distribution.
- Unlike a distribution plot, it doesn't use bins because it places a mark on every single point

```python
sns.rugplot(tips["total_bill"])
```

# Kernel Density Estimation (KDE) Plot
Consists of gaussian distributions (normal distributions such as a bell curve) and rug plots
- Sums up all the gaussian distributions and rug plots points to create a KDE plot

```python
# create a KDE plot and show its a rugplot for total_bill
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```

```python
# create a KDE plot and show its rugplot for tip
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```
