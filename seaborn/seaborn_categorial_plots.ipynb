# Seaborn Categorial Plots

```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
# load the sns tips DataFrame sample
tips = sns.load_dataset("tips")

tips.head()
```

### Bar Plot
Aggregate categorial data into bars

```python
"""
create a barplot of x being the categorial data and y being the numeric data
- the values are estimated by the function in estimator, for this example it
plots based on standard deviation of the data sets
"""
sns.barplot(x=("sex"), y=("total_bill"), data=tips, estimator=(np.std))
```

### Count Plot
The same as the barplot, but the estimator explicity countst the number of occurrences
- Because of this, we only need the x-value

```python
# counts the number of males and females
sns.countplot(x=("sex"), data=tips)
```

### Box Plot (Box and Whisker Plot)
Shows the distribution of the quantative data in a way that shows comparision between variables.
- Shows the median (the horizontal line of the box)
- Shows the range (1st quartile tip subtracted by 4th quartile bottom)
- Shows the outlier datas (outside the quartiles)
- The upper-whisker is the 1st quartile, the upper-box till the horizontal line if the 2nd quartile, the line itself is the median of all the data, the line to the lower-box is the 3rd quartile, and the lower-box to the lower-whisker is the 4th quartile

```python
"""
see the distribution of total_bill per day as a box plot
- the box shows the quartile of the data set
- the horizontal line in the box is the median of the data set
- the whiskers show the extremes (rest of the data set)
- the outlying dots are outlying datas of the data set (irregular data)
- hue tells us whether or not the person was a smoker or not
"""
sns.boxplot(x="day", y="total_bill", data=tips, hue=("smoker"))
```

### Violin Plot
Unlike the Boxplot, violin plot will plot all the data points and even include the outliers. However, this would be a little harder to interpret or read.

```python
"""
create a violin plot for total_bill per day as percentages by sex
- since this is a violin plot, set split=True
    for the hue to combine into a single violin
"""
sns.violinplot(x="day", y="total_bill", data=tips, hue=("sex"), split=True)
```

### Strip Plot
A scatter plot where it scatters the data points.

```python
"""
create a stripplot for the total_bill per day
- jitter=True allows for the data to look better spread out
- hue=("sex") allows us to separate the data by sex
- dodge=True allows us to split the hue instead of overlapping them
"""
sns.stripplot(x="day", y="total_bill", data=tips,
              jitter=True, hue=("sex"), dodge=True)
```

### Swarm Plot
Similar to strip plot, but the points are adjusted so they do not overlap.
- However, sometimes they do not scale properly for large data sets

```python
# use the same distribution of the split plot, but instead as a swarmplot
sns.swarmplot(x="day", y="total_bill", data=tips, hue="sex", dodge=True)
```

```python
# plot both a violin plot and a swarm plot

# plot a violin plot
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex")

# plot a swarm plot with black color
sns.swarmplot(x="day", y="total_bill", data=tips,
              hue="sex", dodge=True, color="black")
```

### Cat Plot (used to be called Factor Plot)
The most general plotting-style from a box, violin, and swarm plot.
- It's mostly better to use the specific plot, but this works too

```python
# plot the total_bill per day as a kind of bar plot
sns.catplot(x="day", y="total_bill", data=tips, kind="bar")
```
