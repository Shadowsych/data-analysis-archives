# Seaborn Regression Plots
- https://seaborn.pydata.org/generated/seaborn.lmplot.html

```python
import seaborn as sns
%matplotlib inline
```

```python
# load the sample tips data set
tips = sns.load_dataset("tips")

tips.head()
```

```python
"""
create a linear model plot with hue as sex, which you can see that
both sexes have pretty similar tip per the value of total_bill
- set the markers as o for male, v for female
- set the size (s) for the markers to 100 from matplotlib
"""
sns.lmplot(x=("total_bill"), y=("tip"), data=tips,
           hue=("sex"), markers=["o", "v"], scatter_kws={"s": 100})
```

```python
"""
create a lmplot separated by the sex column and time row instead of hue
- if you read the plot by rows, you can see the "time" is consistent
- if you read the plot by columns, you can see the "sex" is consistent
"""
sns.lmplot(x=("total_bill"), y=("tip"), data=tips, col=("sex"), row=("time"))
```

```python
"""
create a lmplot separated by the day column and hue of sex
- the aspect ratio is 0.6
- the height is 8
Note: aspect * height gives the width of each facet in inches
"""
sns.lmplot(x=("total_bill"), y=("tip"), data=tips, col=("day"), hue=("sex"),
          aspect=0.6, height=8)
```
