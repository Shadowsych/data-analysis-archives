# MatplotLib Overview

```python
# import matplotlib's Python plotting module
import matplotlib.pyplot as plt
```

```python
# allow for Jupyter Notebook to visualize matplotlib graphs without plt.plot()
%matplotlib inline
```

```python
import numpy as np

# return an Array from 0 to 5 with spacing of 11 items (0.5 spacing)
x = np.linspace(0, 5, 11)
print(x)

# return an Array of x^2
y = x ** 2
print(y)
```

### Functional Matplotlib Plotting (Not Recommended)

```python
# set the x-label
plt.xlabel("X Label")

# set the y-label
plt.ylabel("Y Label")

# set the title
plt.title("Title")

# plot the x, y coordinates as a line graph as a red line
plt.plot(x, y, "r")
```

```python
# create a red-lined sub-plot of 1 row, 2 columns, and as plot #1
plt.subplot(1, 2, 1)
plt.plot(x, y, "r")

# create a blue-lined sub-plot of 1 row, 2 columns, and as plot #2
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

### Object-Oriented Matplotlib Plotting (Better)

```python
# create a blank plot figure
figure = plt.figure()

"""
add the axes (plural of axis) to the figure
[% left of canvas, % bottom of canvas, % width of canvas, % height of canvas]
Note: % left, % bottom has no effect if there's only one graph on the canvas
"""
axes = figure.add_axes([0.1, 0.1, 0.8, 0.8])

# set the labels on the axes
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")

# now plot the values of x, y onto the axes instead
axes.plot(x, y)
```

```python
# create a blank plot figure
figure = plt.figure()

# create an axes in the figure with 10% left and bottom, 80% width and height
axes1 = figure.add_axes([0.1, 0.1, 0.8, 0.8])

# create an axes in the figure with 20% left, 50% bottom, 40% width, 30% height
axes2 = figure.add_axes([0.2, 0.5, 0.4, 0.3])

# plot the larger axes
axes1.set_title("Larger Plot")
axes1.plot(x, y)

# plot the smaller axes
axes2.set_title("Smaller Plot")
axes2.plot(y, x)
```

### Subplots on OOP Matplotlib

```python
# return subplots of 1 row, 2 columns of subplots
figure, axes = plt.subplots(nrows=1, ncols=2)

# the axes is actually a numpy Array of the subplots
print(type(axes))

# however, the figure is still a blank canvas figure
print(type(figure))

# set the axes on the first graph
axes[0].set_title("First Plot")
axes[0].plot(x, y)

# set the axes on the second graph
axes[1].set_title("Second Plot")
axes[1].plot(y, x)

# fix any over-laps that are shown when visualizing the graphs
plt.tight_layout()
```

### Figure Size, Aspect Ratio, and DPI

```python
# create a figure with aspect ratio of (8, 2) with dpi (pixels per inch) of 50
figure = plt.figure(figsize=(8, 2), dpi=50)

# add axes to the figure with full width and column, then plot it
axes = figure.add_axes([0, 0, 1, 1])
axes.plot(x, y)
```

```python
# return subplots of 1 row, 2 columns of subplots with ratio of (8, 2)
figure, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 2))

# plot the axes
axes[0].plot(x, y)
axes[1].plot(y, x)

# fix any over-laps that are shown when visualizing the graphs
plt.tight_layout()
```

### Matplotlib Legends

```python
# create a new blank figure and add axes
figure = plt.figure()
axes = figure.add_axes([0, 0, 1, 1])

# plot x, x^2 and x, x^3 with labels
axes.plot(x, x ** 2, label="X Squared")
axes.plot(x, x ** 3, label="X Cubed")

"""
create a 10% left, 10% bottom legend based on the labels of the plotted axes

there are many locations to plot the legend, but matplotlib's
website highlights some locations. the best (default) is 0.
- https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html
"""
axes.legend(loc=(0.1, 0.1))
```
